```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_excel("Cleaned_Online_Retail.xlsx")
```

```python
# Top 10 countries by transaction count

top_countries = df['Country'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.xticks(rotation=45)
plt.title("Top 10 countries by Transaction Count")
plt.ylabel("Number of Transactions")
plt.xlabel("Country")
plt.show()
```

```python
# Year by year comparison bar chart

uk_sales = df[df['Country'] == 'United Kingdom'].copy()
uk_sales['InvoiceDate'] = pd.to_datetime(uk_sales['InvoiceDate'])
uk_sales['Year'] = uk_sales['InvoiceDate'].dt.year
uk_sales['Revenue'] = uk_sales['Quantity'] * uk_sales['UnitPrice']

yearly_sales = uk_sales.groupby('Year')['Revenue'].sum().reset_index()

plt.figure(figsize=(10,5))
sns.barplot(x='Year', y='Revenue', data=yearly_sales, palette='Reds_d')
plt.title("Year By Year Sales of UNITED KINGDOM")
plt.xlabel("Year")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Year- 2010 Country wise Revenue excluding United Kingdom

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

df_2010 = df[(df['InvoiceDate'].dt.year == 2010) & (df['Country'] != 'United Kingdom')].copy()

df_2010['Revenue'] = df_2010['Quantity'] * df_2010['UnitPrice']

country_revenue = df_2010.groupby('Country')['Revenue'].sum().reset_index()

country_revenue = country_revenue.sort_values(by='Revenue', ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x='Country', y='Revenue', data=country_revenue, palette='viridis')
plt.title("Year-2010 Country wise Revenue excluding United Kingdom")
plt.xlabel("Country")
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout
plt.show()
```

```python
# Year 2011 Country wise revenue excluding United Kingdom

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

df_2011_no_uk = df[(df['InvoiceDate'].dt.year == 2011) & (df['Country'] != 'United Kingdom')].copy()

df_2011_no_uk['Revenue'] = df_2011_no_uk['Quantity'] * df_2011_no_uk['UnitPrice']

country_revenue_2011 = df_2011_no_uk.groupby('Country')['Revenue'].sum().reset_index()

country_revenue_2011 = country_revenue_2011.sort_values(by='Revenue', ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x='Country', y='Revenue', data=country_revenue_2011, palette='magma')
plt.title('Year-2011 Country wise revenue excluding United Kingdom')
plt.xlabel('Country')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
